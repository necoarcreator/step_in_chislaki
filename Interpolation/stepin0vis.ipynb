```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return np.sin(x/3 + np.exp(np.sin(x/3) * np.sin(x/3)))
```

```python
a = 0.
b = 10.
n = 20
```

```python
x = np.linspace(a, b, n)
y = f(x)
```

```python
string_1 = []
string_2 = []
N_1 = 0
N_2 = 0
i = 0
```

```python
with open('Interp_New.txt', 'r') as file:
    for item in enumerate(file):
        N_1 += 1
        string_1.append(item[1])
```

```python
with open('Interp_Lag.txt', 'r') as file_2:
    for item in enumerate(file_2):
        N_2 += 1
        string_2.append(item[1])
```

```python
for k in range(N_1):
    string_1[k] = string_1[k].split('/')
```

```python
for k in range(N_2):
    string_2[k] = string_2[k].split('/')
```

```python
x_new = np.zeros(N_1)
y_new = np.zeros(N_1)
x_lag = np.zeros(N_2)
y_lag = np.zeros(N_2)
for k in range(N_1):
    x_new[k] = string_1[k][0]
    y_new[k] = string_1[k][1]
    
for k in range(N_2):
    x_lag[k] = string_2[k][0]
    y_lag[k] = string_2[k][1]
```

```python
plt.scatter(x, y, label = 'Original spline')
plt.plot(x_lag[5:], y_lag[5:], lw = 2.0, color = 'red', label = 'Lagrange')
plt.plot(x_new[2:99], y_new[2:99], lw = 2.0, ls = '--', color = 'green', label = 'Newton')
plt.title('Interpolation of $f(x) = sin(x/3 + exp^2(x/3))$ with usual nodes')
plt.xlabel('$0 < x < 10$')
plt.ylabel('$-1 < f(x) < +1$')
plt.grid(color = 'brown', ls = ":", lw = 0.5)
plt.text(12, 0, "Computational complexity is $O(N^3)$", fontsize =12)
plt.legend()
```

```python
x = np.zeros(n + 1)


for i in range(n + 1):
    x[i] = (a + b) / 2 + (b - a) * np.cos(np.pi * (2 * i + 1) / (2 * n)) / 2;
    
y = f(x)
```
